## Spotify Insights

# Load libraries and data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from utils import (
    top_artists_by_time,
    top_artists_by_streams,
    top_songs,
    days
)
```

# Load and preprocess data

```python
df_history_0 = pd.read_json("data/StreamingHistory_music_0.json")
df_history_1 = pd.read_json("data/StreamingHistory_music_1.json")
df_history = pd.concat([df_history_0, df_history_1], ignore_index=True)
df_history["minPlayed"] = df_history["msPlayed"] / 60000
df_history["endTime"] = pd.to_datetime(df_history["endTime"])
df_history["date"] = df_history["endTime"].dt.date
df_history["time"] = df_history["endTime"].dt.time
```

# Top artists by listening time

```python
top_artists_by_time(df_history)
```

# Top artists by streams

```python
top_artists_by_streams(df_history)
```

```python
top_songs(df_history)
```
